```python
import cv2
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
# get image from web for demostration
!wget -q "https://cdn.theplaylist.net/wp-content/uploads/2022/06/11130427/Ben-Affleck-Batman.jpg" -O 'demo.jpg'
```

```python
original_img = cv2.imread('demo.jpg')
cv2_imshow(original_img)
```

```python
gray_scale_img = cv2.cvtColor(original_img, cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(10, 10))
plt.imshow(gray_scale_img, cmap='gray')
```

```python
# bluring the image using gaussian blur with kerenel 5x5
blur_image = cv2.GaussianBlur(gray_scale_img, (5, 5), 0)

# sketching edge
edge_img= cv2.Canny(gray_scale_img, 135, 175)
plt.figure(figsize=(10, 10))
plt.imshow(edge_img, cmap='gray')
```

```python
pencil_sketch = cv2.adaptiveThreshold(gray_scale_img, 255,
    cv2.ADAPTIVE_THRESH_MEAN_C,
    cv2.THRESH_BINARY, 9, 7)  # thickness of line of edge, density of no of lines

plt.figure(figsize=(10, 10))
plt.imshow(pencil_sketch, cmap='gray')
plt.axis(False)
```
